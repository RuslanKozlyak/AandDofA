```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict 
```

```python
N = 100
V = 200
b = np.zeros(shape=(N,N))
upper_triangle = np.triu_indices(N,1)

indexes = np.arange(0, len(upper_triangle[0]))
indexes = np.random.choice(indexes, V, replace=False)

for i in indexes:
    idx1 = upper_triangle[0][i]
    idx2 = upper_triangle[1][i]
    b[idx1,idx2] = 1

adj_matrix = (b + b.T)
adj_matrix
```

```python
adj_list = defaultdict(list)

for i in range(N):
    for j in range(N):
        adj_list[i].append(j) if adj_matrix[i, j] == 1 else next

adj_list
```

```python
class GraphVisualization:
    def __init__(self, temp):
        self.visual = temp
    def visualize(self):
        G = nx.Graph()
        plt.figure(figsize=(20,15))
        plt.title(f'My random graph', fontsize=25)
        G.add_edges_from(self.visual)
        nx.draw_networkx(G)
        plt.show()


lst = []
for k, v in adj_list.items():
    lst.extend([[k, i] for i in v])
G = GraphVisualization(lst)
G.visualize()

print(' --- Random adjacency matrix rows ---')
print(adj_matrix[np.random.randint(0, 99, size=2)])
print(' ')
print(' --- Random adjacency list rows ---')
k_rand = np.random.randint(0, 99, size=2) 
print(f'{k_rand[0]} -> {adj_list[k_rand[0]]} \n{k_rand[1]} -> {adj_list[k_rand[1]]}')
```
